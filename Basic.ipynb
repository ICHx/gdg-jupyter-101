How to read a CSV
=========
<br/>
<br/>
<br/>
<br/>

Reading CSV is very simple by calling `pd.read_csv`.

`df.head` is to output the first few rows of a data frame.

A data frame represents a tabular data structure which contains `data`, `columns` and `indices`.


<br/>
<br/>

```python
import pandas as pd
from dateutil.parser import parse
df = pd.read_csv('pollution.csv', na_values = '-')
df['DateTime'] = df['DateTime'].apply(lambda x: parse(x))
df = df.sort_values(by='DateTime', ascending=False)
df = df.reset_index(drop=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df2 = df[df["StationName"] == "Mong Kok"][["NO2", "PM10"]]
df2["Hour"] = df["DateTime"].apply(lambda x: x.hour) 
df2["Day"] = df["DateTime"].apply(lambda x: x.day) 
df2[df2["Day"] == 10].plot(x='Hour', y=['NO2', 'PM10'], title='NO2 & PM10', figsize=(20,20))
df2[df2["Day"] == 11].plot(x='Hour', y='NO2', figsize=(20,20), title='NO2 Only')
```

```python
import pandas as pd
from dateutil.parser import parse
df = pd.read_csv('pollution.csv',  na_values = '-')
df['DateTime'] = df['DateTime'].apply(lambda x: parse(x))
df['Day'] = df['DateTime'].dt.day
df = df.sort_values(by=['DateTime', 'NO2'], ascending=[False, False])
df2 = df[(df['Day'] == 10) & (df['StationName'] == 'Mong Kok')]
df.dtypes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df2['Hour'] = df['DateTime'].dt.hour
df2.plot(x='Hour', y='NO2',figsize=(20,20))
```

```python
api = pd.read_csv('aqih.csv',  na_values = '-')
api['DateTime'] = api['DateTime'].apply(lambda x: parse(x))
merged = pd.merge(df, api, how='outer', on=['DateTime', 'StationName'])
new_merged = merged[['DateTime', 'Day', 'StationName', 'NO2', 'aqhi']]
new_merged['is_day'] = (merged['DateTime'].dt.hour >= 6) & (merged['DateTime'].dt.hour <= 18)
new_merged.head(n=100)
```

```python
plt.figure(figsize=(20, 20))
mk = new_merged[new_merged['StationName'] == 'Mong Kok']
plt.title('NO2 & API')
plt.plot(mk['DateTime'], mk['aqhi'], color='red')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from dateutil.parser import parse
df = pd.read_csv('aqih.csv', na_values='-')
df['DateTime'] = df['DateTime'].apply(lambda x: parse(x))
print(df['StationName'].unique())
df = df[(df['DateTime'].dt.day == 11) & (df['StationName'] == 'Tai Po')]
plt.figure(figsize=(10, 10))
plt.title('API')
plt.subplot(1, 1, 1)
plt.plot(df['DateTime'], df['aqhi'], color='red')
plt.plot(df['DateTime'], df['aqhi'] + 1, color='blue')
plt.show()
plt.figure(figsize=(20, 20))
plt.title('API Scatter')
plt.scatter(df['DateTime'].dt.hour,df['aqhi'], color='green')
plt.show()
```

```python
df = pd.read_csv('aqih.csv', na_values='-')
df['DateTime'] = df['DateTime'].apply(lambda x: parse(x))
df = df.sort_values(by='DateTime')
df = df[df['DateTime'].dt.day == 10]
df3 = df[df['StationName'] == 'Tai Po'][['DateTime', 'aqhi']]
df4 = df[df['StationName'] == 'Mong Kok'][['DateTime', 'aqhi']]
df5 = df[df['StationName'] == 'Sha Tin'][['DateTime', 'aqhi']]

plt.title('API Tai Po vs Mong Kok')
plt.plot(df3['DateTime'].dt.hour, df3['aqhi'], color='green')
plt.plot(df4['DateTime'].dt.hour, df4['aqhi'], color='blue')
plt.plot(df5['DateTime'].dt.hour, df5['aqhi'], color='red')
plt.show()
print(df['StationName'].unique())
```

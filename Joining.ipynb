```python
import pandas as pd
from dateutil.parser import parse
df = pd.read_csv('pollution.csv', na_values = '-')
df['DateTime'] = df['DateTime'].apply(lambda x: parse(x))
df[df['CO'].notna()]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df2 = pd.read_csv('aqih.csv')
df2['DateTime'] = df2['DateTime'].apply(lambda x: parse(x))
df2.head()
```

```python
df3 = df[['DateTime', 'StationName', 'PM2.5']]
df4 = df2[['DateTime', 'StationName', 'aqhi']]
df3 = df3[(df3['DateTime'].dt.day == 10) & (df3['StationName'] == 'Central/Western')]
df4 = df4[(df4['DateTime'].dt.day == 10)& (df4['StationName'] == 'Central/Western')]
df4 = pd.merge(df3, df4, how='outer', on=['DateTime', 'StationName'])
df4.plot(x='DateTime', y=['PM2.5', 'aqhi'])
df4
```
